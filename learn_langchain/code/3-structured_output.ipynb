```python
import os
from dotenv import load_dotenv
from langchain_groq import ChatGroq
```

```python
load_dotenv()
```

```python
model = ChatGroq(
    model="qwen/qwen3-32b",
    reasoning_effort="none"
)
```

```python
# Define the structured output schema using Pydantic
from pydantic import BaseModel,Field

class Movie(BaseModel):
    title: str = Field(..., description="The title of the movie")
    year: int = Field(..., description="The release year of the movie")
    director: str = Field(..., description="The director of the movie")
    genre: str = Field("Unknown", description="The genre of the movie")
```

```python
model_with_structured_output = model.with_structured_output(Movie)
```

```python
response = model_with_structured_output.invoke("Please provide details of the movie 'Inception'.")
response
```
